```python
#Soal 1.1

%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(0,10)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,'b',x,z,'r')
plt.xlabel('radians')
plt.ylabel('value')
plt.title('plotting demonstration')
plt.legend(['sin', 'cos' ])
plt.grid()
```

```python
#soal 1.2

beratEmas = 25
hargaSatuGram = 650000
kenaikanHargaEmasSatuGram = 685000

#dalamrupiah(Rp)
KeuntunganMasGerard = ((beratEmas * kenaikanHargaEmasSatuGram) - (beratEmas * hargaSatuGram))
print("jadi keuntungan mas gerard dalam rupiah adalah =  Rp",KeuntunganMasGerard)

#dalampersen(%)
berapaPersen = float(KeuntunganMasGerard/(kenaikanHargaEmasSatuGram * beratEmas)) * 100 
print("jadi persentase keuntungan mas gerard  adalah  = ", berapaPersen,'%\n')

print('sabar sek mas gerard beli emas lagi/ wait a moment Mr gerard buy gold again')
print('\t\t\t\t hehe\n')

#mas gerard banyak duit jir, beli lagi doi 

beliEmas = 15
#total emas yang dipunya 
dpunya = 40
hargaBaru = 715000

red = float((beratEmas * hargaSatuGram) + (beliEmas * kenaikanHargaEmasSatuGram))
blue= float(dpunya * hargaBaru)
green = float(blue - red)
print('jadi keuntungan kedua mas gerard adalah = Rp',green)

#dalam persen
gerardSugih = (green/blue) * 100
print('persentasi keuntungan mas gerard secara total adalah =',gerardSugih,'%')

```

```python
#soal 1.3

saldoAwal = 200000000
bunga =10/100
tujuan = 400000000
mulaiTahun = 0

while saldoAwal < tujuan:
    mulaiTahun += 1
    hitung = saldoAwal * (1 + bunga)
    hasil = hitung - saldoAwal
    saldoAwal+= hasil
    
print('membutuhkan waktu', mulaiTahun,'tahun, ''untuk mengasilkan', round(hitung))
```
